## Acessar dados do GFS via OPeNDAP

O que é GFS?

[**Global Forecast System (GFS)**](https://www.ncdc.noaa.gov/data-access/model-data/model-datasets/global-forcast-system-gfs): O Global Forecast System (GFS) é um modelo de previsão do tempo produzido pelos Centros Nacionais de Previsão Ambiental (NCEP). Dezenas de variáveis atmosféricas e terrestres estão disponíveis por meio desse conjunto de dados, desde temperaturas, ventos e precipitação até umidade do solo e concentração de ozônio atmosférico. O globo inteiro é coberto pelo GFS em uma resolução horizontal de base de 18 milhas (28 quilômetros) entre os pontos da grade, que é usada pelos analistas operacionais que prevêem o tempo até 16 dias no futuro. A resolução horizontal cai para 44 milhas (70 quilômetros) entre o ponto da grade para previsões entre uma e duas semanas.

O que é OPeNDAP? 

[**OPeNDAP**](https://www.opendap.org/): O Open-source Project for a Network Data Access Protocol(OPeNDAP) significa "Projeto de código aberto para um protocolo de acesso a dados de rede". OPeNDAP é o nome de uma organização sem fins lucrativos e o nome comumente usado de um protocolo que a organização OPeNDAP desenvolveu. O protocolo DAP2 fornece um meio disciplinar neutro de solicitar e fornecer dados pela World Wide Web. O objetivo é permitir que os usuários finais, sejam eles quem forem, acessem imediatamente quaisquer dados de que necessitem de uma forma que possam usar, ao mesmo tempo que usam aplicativos que já possuem e com os quais estão familiarizados. No campo da oceanografia, o OPeNDAP já ajudou a comunidade de pesquisa a fazer progressos significativos nesse sentido. Em última análise, espera-se que OPeNDAP seja um componente fundamental de sistemas que fornecem interoperabilidade máquina a máquina com significado semântico em um ambiente altamente distribuído de conjuntos de dados heterogêneos. A organização OPeNDAP existe para desenvolver, implementar e promulgar o protocolo OPeNDAP. Ele apresenta os resultados de seu trabalho gratuitamente ao público, com a esperança de que seja útil em muitas disciplinas e facilite o compartilhamento e o acesso aos seus fluxos de dados.

* uma estrutura que simplifica todos os aspectos da rede de dados científicos.
* fornece software que torna os dados locais acessíveis a locais remotos, independentemente do formato de armazenamento local.
* também fornece ferramentas para transformar aplicativos existentes em clientes OPeNDAP (ou seja, permitindo que eles acessem remotamente os dados servidos por OPeNDAP).
* O software OPeNDAP está disponível gratuitamente.

## Obtenção dos dados

```python
# link do opendap para os dados do GFS
gfs_opendap = 'https://nomads.ncep.noaa.gov/dods/gfs_0p25_1hr/gfs20210324/gfs_0p25_1hr_00z'
```


O `Xarray` é quem faz a mágica de acessar o servidor através do link. Para abrir qualquer arquivo de dados no formato Netcdf (ou até mesmo grib e raster), basta usar a função `.open_dataset()`.

```python
# Abra os arquivos e instancie para gfs
gfs = xr.open_dataset(gfs_opendap)
```

```python
gfs
```

Utilizaremos das 250 variáveis a variável  Temperatura do Ar em 2 metros

```python
# Selecione a variável de interesse
temperatura = gfs['tmp2m']
```

```python
# Visualize o dataset de temperatura
temperatura
```

Agora, vamos visualizar o primeiro campo global de `Temperatura` do **GFS** com `Proplot`

```python
# Plote o primeiro campo de Temperatura
fig, ax = plot.subplots(axwidth=4.5, tight=True,
                      proj='pcarree', proj_kw={'lon_0': 0},)

ax.format(land=False, coast=True, innerborders=True, borders=True,
          large='15px', labels=True,
          geogridlinewidth=0,
          abcloc='ur',)

map1 = ax.contourf(temperatura['lon'], temperatura['lat'],
                   temperatura[0, :, :]-273.15,
                   levels=plot.arange(-30, 30, 5), cmap='coolwarm',
                   extend='both')

ax.colorbar(map1, loc='b', shrink=0.9, label='Temperatura $\degree$C')

plot.show();
```
